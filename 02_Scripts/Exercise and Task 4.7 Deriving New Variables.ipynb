### Q4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# Creating a dataframe with the count of values for the column 'order_hour_of_day'

counts = df_ords_prods_merged['order_hour_of_day'].value_counts().rename_axis('hour').reset_index(name='counts')
```

```python
counts
```

```python
# Creating a list showing the percentiles for the value counts of the column 'order_hour_of_day'

orders = list(df_ords_prods_merged['order_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
orders
```

#### Lists: I created three lists for the three different labels: Most orders, Average orders, and Fewest orders"

```python
# Fewest orders list

few_ords = list(counts.loc[counts['counts'] <= orders[0]]['hour'])
```

```python
# Average orders list

avg_ords = list(counts[(counts['counts'] > orders[0]) & (counts['counts'] < orders[2])]['hour'])
```

```python
# Most orders list

most_ords = list(counts.loc[counts['counts'] >= orders[2]]['hour'])
```

#### Index: I created three indexs for the three different labels: Most orders, Average orders, and Fewest orders"

```python
index_few_ords = df_ords_prods_merged['order_hour_of_day'].isin(few_ords)
```

```python
index_avg_ords = df_ords_prods_merged['order_hour_of_day'].isin(avg_ords)
```

```python
index_most_ords = df_ords_prods_merged['order_hour_of_day'].isin(most_ords)
```

#### Create a new column containing these labels called busiest_period_of_day.

```python
df_ords_prods_merged.loc[index_few_ords,'busiest_period_of_day'] = 'Fewest orders'
```

```python
df_ords_prods_merged.loc[index_avg_ords,'busiest_period_of_day'] = 'Average orders'
```

```python
df_ords_prods_merged.loc[index_most_ords,'busiest_period_of_day'] = 'Most orders'
```
